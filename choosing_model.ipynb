```
#importing pandas
import pandas as pd  
```

```
#reading csv file. Remember that each time you start the environment, you need to upload the file diabetes_data.csv again
df = pd.read_csv("diabetes_data.csv")
```

```
#checking the dataframes's shape
df.shape
```

```
df.head()
```

```
#transforming the data in numpy.ndarray
x = df.iloc[:,:-1].values
y = df.iloc[:,8].values
```

```
#importing MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

#normalizing the data
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

```
#importing train_test_split
from sklearn.model_selection import train_test_split

#making the division between training data (70%) and testing data (30%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```
#importing the algorithm KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier

#instantiation of the class
KNN_clf = KNeighborsClassifier(n_neighbors=5)
KNN_clf.fit(x_train, y_train)

#predicting
y_pred = KNN_clf.predict(x_test)

#importing classification report to find the accuracy
from sklearn.metrics import  classification_report
KNN_accuracy = round(classification_report(y_test, y_pred, output_dict = True)['accuracy'],2)
print(KNN_accuracy)
```

```
#decision tree (the same steps as before, but with another algorithm)
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

tree_clf = DecisionTreeClassifier(random_state=1)

tree_clf = tree_clf.fit(x_train, y_train)

y_pred = tree_clf.predict(x_test)

tree_accuracy = round(classification_report(y_test, y_pred, output_dict = True)['accuracy'],2)
print(tree_accuracy)
```

```
from sklearn.neural_network import MLPClassifier

MLP_clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5,
10), random_state=1)

MLP_clf = MLP_clf.fit(x_train, y_train)

y_pred = MLP_clf.predict(x_test)

MLP_accuracy = round(classification_report(y_test, y_pred, output_dict = True)['accuracy'],2)
print(MLP_accuracy)
```

```
#exporting the model
import joblib

filename="prediction_model.sav"
joblib.dump(MLP_clf,filename)
```
